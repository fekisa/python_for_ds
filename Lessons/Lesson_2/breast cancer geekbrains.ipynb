Видим, что cellularity можно перевести в количественный признак, так как ее значения количественно соотносятся друг с другом

```python
print(df['cellularity'].unique())
df_cel = df['cellularity'].replace(['High', 'Moderate', 'Low'],[3,2,1]).fillna(0)
df_cel
```

Выберем колонки с категориальными признаками, показывающими типы мутаций в генах

```python
df_mut = df[[col for col in df.columns if '_mut' in col]].astype('str')
df_mut
```

Видим, что в колонках очень много нулей. Посчитаем кол-во нулей в каждой колонке, отсортируем гены по этому числу и выведем их для всех генов специальной функцией

```python
display(HTML(  df_mut.apply(lambda x: x.value_counts()).transpose()[['0']].sort_values(by=['0']).to_html()  ))
```

Сделаем датафрейм с колонками, в которых меньше 1700 нулей

```python
df_mut_select_cols = df_mut.apply(lambda x: x.value_counts()).transpose()[['0']]
df_mut_1700_cols = df_mut_select_cols[df_mut_select_cols['0'] < 1700].index.values
df_mut_1700 = df[df_mut_1700_cols]
df_mut_1700
```

Там, где нулей больше 1700 заменим мутации на 1, так мы получим признаки, которые будут показывать, что в этих генах есть хотя бы одна любая мутация

```python
df_mut_1699_cols = df_mut_select_cols[df_mut_select_cols['0'] >= 1700].index.values
df_mut_1699 = df[df_mut_1699_cols]

for col in df_mut_1699_cols:
    df_mut_1699.loc[df_mut_1699[col] != '0', col] = '1'

df_mut_1699 = df_mut_1699.astype('int')
print(df_mut_1699.info())
df_mut_1699
```

### Посмотрим на гены с частыми мутациями

Смотрим на отдельные мутации и понимаем, что бывают случаи когда у одного человека сразу несколько мутаций. Поэтому хорошо бы для таких случаев не делать новые колонки для каждого сочетания мутаций, а сделать бинарные признаки, содержащие по одной мутации

```python
display(HTML(df_mut_1700[df_mut_1700['pik3ca_mut'] != '0'][['pik3ca_mut']].to_html()))
```

Функция, которая для каждого гена делает колонки со всеми его возможными единичными мутациями и ставит 1, если у человека есть мутация

```python
df_mut_1700_final = pd.DataFrame()
def make_new_mut(mut_name):
    print(mut_name)
    single_mut = []
    for mut in df_mut[mut_name].unique():
        for mut1 in mut.split():
            single_mut.append(mut1)
    single_mut = list(set(single_mut))
    
    for mut in single_mut:
        #print(mut)
        for indx, row  in df_mut.iterrows():
            if mut in row[mut_name]: 
                df_mut_1700_final.loc[df_mut.index[indx], mut_name+'='+mut] = 1
            else: df_mut_1700_final.loc[df_mut.index[indx], mut_name+'='+mut] = 0
```

```python
for m in df_mut_1700_cols:
    make_new_mut(m)
    
df_mut_1700_final
```
